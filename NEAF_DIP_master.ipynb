## **1. Histogram Equalization**
Histogram is the method to enhance image contrast.  
Please show the image grayscale histogram and do the equlization, and show the image result and histogram equalization result.  

<img src="md_picture/1.png" width="60%" heught="60%">

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_ori = cv2.imread('input/lena.bmp', cv2.IMREAD_GRAYSCALE) # Read image as gray.
plt.figure()
plt.imshow(img_ori,cmap='gray')
equal = cv2.equalizeHist(img_ori)
plt.figure()
plt.imshow(equal,cmap='gray')
"""
Your code here
"""

histogram = cv2.calcHist([img_ori], [0], None, [256], [0, 256])

x_range = np.arange(256)
plt.figure()
plt.title('Histogram origin')
plt.xlabel('grayscale')
plt.ylabel('pixal')
plt.bar(x_range, histogram.ravel(), color='green')
plt.xlim([0, 256])


plt.figure()
histogram_equal=cv2.calcHist([equal], [0], None, [256], [0, 256])
plt.title('Histogram equalized')
plt.xlabel('grayscale')
plt.ylabel('pixal')
plt.bar(x_range, histogram_equal.ravel(), color='green')#.ravel一維展開
plt.show()
# save result
# cv2.imwrite('output/img_ori.bmp', img_ori)
# cv2.imwrite('output/img_equa.bmp', img_equa)
```

## **4. HSV filter**
This method could extract the the interested part by color.  
Please crop the yellow or blue boxes and its surrounding in the image.

<img src="md_picture/5.png" width="30%" heught="30%">

### Steps:
1. Transfer color to HSV.
2. Setting the color range and filter the image. (output the color image in the setting range)
3. Binary the image, find the left-top and right-bottom coordinate.
4. Crop the image by the coordinate.  

- Noted: Blur, histogram equalization methods could be used if needed.  

```python
# Yellow box.
yellow = cv2.imread('input/yellow.jpg')
#yellow = cv2.imread('input/lena.bmp')
plt.figure()
yellow_rgb = cv2.cvtColor(yellow, cv2.COLOR_BGR2RGB)
plt.imshow(yellow_rgb)
plt.axis('off')

plt.figure()
# Step 1. Transfer color to HSV
yellow_hsv = cv2.cvtColor(yellow, cv2.COLOR_BGR2HSV)
# Step 2. Setting the color range and filter the image.
lower_yellow = np.array([30,10, 150])
upper_yellow = np.array([60, 255, 255])
range = cv2.inRange(yellow_hsv, lower_yellow, upper_yellow)

result_img = cv2.bitwise_and(yellow, yellow, mask=range)
# plt.imshow(result_img)
# plt.axis('off')
plt.figure()
result_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
plt.imshow(result_rgb)
plt.axis('off')
# Step 3. Binary the image, find the left-top and right-bottom coordinate (x, y, w, h) or (x1, x2, y1, y2).
x, y, w, h = cv2.boundingRect(range)
# Step 4. Crop the image by the coordinate.
plt.figure()
crop_edge = 20  # surrounding 20 pixels 
cropHSV_yellow = yellow[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.
cropHSV_yellow =cv2.cvtColor(cropHSV_yellow, cv2.COLOR_BGR2RGB)
plt.imshow(cropHSV_yellow)
plt.axis('off')


# save result
cv2.imwrite('output/cropHSV_yellow.jpg', cropHSV_yellow)
```

```python
# Blue box.
blue = cv2.imread('input/blue.jpg')
# plt.figure()
# blue_rgb = cv2.cvtColor(blue, cv2.COLOR_BGR2RGB)
# plt.imshow(blue_rgb)
# plt.axis('off')

# Step 1. Transfer color to HSV
blue_hsv = cv2.cvtColor(blue, cv2.COLOR_BGR2HSV)
# Step 2. Setting the color range and filter the image.
lower_blue = np.array([70,100, 0])
upper_blue = np.array([90, 255, 100])
range = cv2.inRange(blue_hsv, lower_blue, upper_blue)

result_img = cv2.bitwise_and(blue, blue, mask=range)
plt.imshow(result_img)
plt.axis('off')
# Step 3. Binary the image, find the left-top and right-bottom coordinate (x, y, w, h) or (x1, x2, y1, y2).
x, y, w, h = cv2.boundingRect(range)
# Step 4. Crop the image by the coordinate.
crop_edge = 20  # surrounding 20 pixels 
cropHSV_blue = blue[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.
cropHSV_blue =cv2.cvtColor(cropHSV_blue, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(cropHSV_blue)
plt.axis('off')
# save result
cv2.imwrite('output/cropHSV_blue.jpg', cropHSV_blue)
```

Use package to find the contours in the image.

Practice:
1. Find the pad and probemark contours in the pad series images.
2. Calculate the countours area.

<img src="md_picture/7.png" width="60%" heught="60%">

```python
pad_task = cv2.imread('input/binary_task1.bmp')
pad_task1= cv2.cvtColor(pad_task, cv2.COLOR_BGR2GRAY)
# Find the pad and probemark contours and draw.

"""
Your code here
"""
thres=128
_, binary_img = cv2.threshold(pad_task1, thres, 255, cv2.THRESH_BINARY)
kernel = np.ones((15, 15), np.uint8)
dilated_img = cv2.dilate(binary_img, kernel, iterations=1)#膨脹
plt.figure()
plt.imshow(dilated_img, cmap='gray')
plt.axis('off')
eroded_img = cv2.erode(dilated_img, kernel, iterations=1)#侵蝕

contours, _ = cv2.findContours(eroded_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
pad_task1_cont = np.copy(pad_task)
cv2.drawContours(pad_task1_cont, contours, -1, (0, 0, 255), 2)
plt.imshow(cv2.cvtColor(pad_task1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.figure()
plt.imshow(cv2.cvtColor(eroded_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.figure()
plt.imshow(cv2.cvtColor(pad_task1_cont, cv2.COLOR_BGR2RGB))
plt.axis('off')
for i in contours:
    pad_area = cv2.contourArea(i)

# save and print result
cv2.imwrite('output/pad_task1_cont.bmp', result)
print('The pad area is ', pad_area)
# print('The probe mark area is ', probemark_area)
```
